# Trebalo bi na slucaja nacin da odabewremo centroide, pa cemo izabrane postaviti u tabelu sa pd.concat 

```python
def random_centroids(data , k):
    list = []
    for i in range(k):
        centroids = data.apply(lambda x: float(x.sample()))
        list.append(centroids)
    return pd.concat(list, axis = 1)
```

```python
def new_centroids( data , labels, k):
    return data.groupby(labels).apply(lambda x : np.exp(np.log(x).mean())).T
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from IPython.display import clear_output
```

PCA cu koristiti da bih sa vise dimenzija presla na 2D prikazivanje

```python
def plot_clusters(data , labels, list1 , iteration):

    pca = PCA(n_components = 2)
    data_2d = pca.fit_transform(data)
    centroids_2d = pca.transform(list1.T)
    print (centroids_2d)
    print (centroids_2d[:,0])
    clear_output(wait = True)
    plt.title(f'Iterarion {iteration}')
    plt.scatter(x=data_2d[:,0], y=data_2d[:,1] , c = labels)
    plt.scatter(x=centroids_2d[:,0] ,  y= centroids_2d[:,1] )
    plt.show()
```

```python
max_iterations = 100
k = 3
list1 = random_centroids(data, k)
old_centroids = pd.DataFrame()
iteration = 1
while iteration < max_iterations and not list1.equals(old_centroids):
        old_centroids = list1
        
        
        labels = get_labels(data, list1)
        list1 = new_centroids(data, labels, k)
        print (labels.value_counts())
        
        plot_clusters(data, labels, list1 , iteration)
        iteration +=1
        
```

```python
from sklearn.cluster import KMeans
Sum_of_squared_distances = []
K = range(1,10)
for num_clusters in K :
 kmeans = KMeans(n_clusters=num_clusters)
 kmeans.fit(data)
 Sum_of_squared_distances.append(kmeans.inertia_)
plt.plot(K,Sum_of_squared_distances,'bx-')
plt.xlabel('Values of K') 
plt.ylabel('Sum of squared distances/Inertia') 
plt.title('Elbow Method For Optimal k')
plt.show()
```
